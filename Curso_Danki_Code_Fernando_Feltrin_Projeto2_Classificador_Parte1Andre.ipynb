````python
METRICS = [  # 📊 Lista de métricas usadas para avaliar o desempenho do modelo de machine learning.

    BinaryAccuracy(name='accuracy'),  # ✅ Mede a acurácia do modelo em classificações binárias (0 ou 1).
                                      #    - Calcula a proporção de previsões corretas sobre o total de amostras.

    TruePositives(thresholds=0.5, name='tp'),  # 🎯 Conta quantas vezes o modelo classificou corretamente um caso positivo (1).
                                               #    - Exemplo: Classificar corretamente um paciente como doente.

    TrueNegatives(thresholds=0.5, name='tn'),  # ✅ Conta quantas vezes o modelo classificou corretamente um caso negativo (0).
                                               #    - Exemplo: Classificar corretamente um paciente como saudável.

    FalsePositives(thresholds=0.5, name='fp'),  # 🚨 Conta os falsos positivos (quando o modelo previu 1, mas o correto era 0).
                                                #    - Exemplo: Diagnosticar um paciente saudável como doente.

    FalseNegatives(thresholds=0.5, name='fn'),  # ⚠️ Conta os falsos negativos (quando o modelo previu 0, mas o correto era 1).
                                                #    - Exemplo: Diagnosticar um paciente doente como saudável.

    PrecisionAtRecall(recall=0.5, name='precision'),  # 🎯 Mede a precisão do modelo para um nível específico de recall (sensibilidade).
                                                      #    - Precisão = (TP) / (TP + FP).
                                                      #    - Útil para evitar falsos positivos em aplicações críticas.

    SensitivityAtSpecificity(0.5, name='sensitivity'),  # 🔬 Mede a sensibilidade (recall) para um valor específico de especificidade.
                                                        #    - Sensibilidade = (TP) / (TP + FN).
                                                        #    - Útil para detectar o máximo de casos positivos.

    SpecificityAtSensitivity(sensitivity=0.5, name='specificity'),  # 🛡️ Mede a especificidade para um nível específico de sensibilidade.
                                                                    #    - Especificidade = (TN) / (TN + FP).
                                                                    #    - Útil para minimizar falsos positivos.

    Recall(name='recall')  # 🔄 Mede a capacidade do modelo de identificar corretamente os positivos.
                           #    - Recall = (TP) / (TP + FN).
                           #    - Importante para modelos onde perder um caso positivo é crítico (exemplo: diagnóstico de doenças).

]  # 🔢 Essa lista será usada ao compilar o modelo, permitindo acompanhar diferentes métricas durante o treinamento e a avaliação.
````

````python
data = datasets.load_breast_cancer()  # 🩺 Carrega o conjunto de dados Breast Cancer (Câncer de Mama) do Scikit-Learn.
                                      #    - Esse dataset contém informações sobre tumores mamários, usados para prever se são benignos (0) ou malignos (1).
                                      #    - Possui 30 características numéricas extraídas de imagens de tumores, como:
                                      #      - Raio médio
                                      #      - Textura média
                                      #      - Perímetro médio
                                      #      - Área média, entre outras.
                                      #    - Esse conjunto de dados é amplamente utilizado para problemas de classificação binária em machine learning.
````

````python
X = pd.DataFrame(data=data.data,  # 🗂️ Converte os dados do dataset Breast Cancer em um DataFrame do Pandas.
                 columns=data.feature_names)  # 🏷️ Define os nomes das colunas com as características do dataset.

# 🔹 Explicação:
#    - `data.data` contém os valores numéricos das características do tumor.
#    - `data.feature_names` fornece os nomes das 30 características (ex.: 'mean radius', 'mean texture', etc.).
#    - Criar um DataFrame facilita a manipulação e análise dos dados antes do treinamento do modelo.
````

````python
print(X.head())  # 🔍 Exibe as 5 primeiras linhas do DataFrame X.
                 #    - Cada linha representa uma amostra (tumor).
                 #    - Cada coluna representa uma característica do tumor (ex.: raio médio, textura média, etc.).
                 #    - Útil para visualizar a estrutura dos dados antes da modelagem.
````

````python
print(X.info())  # 📊 Exibe um resumo informativo do DataFrame X.
                 #    - Mostra o número total de entradas (linhas) e colunas.
                 #    - Exibe os tipos de dados de cada coluna (ex.: float64, int64).
                 #    - Indica se há valores ausentes (NaN) no conjunto de dados.
                 #    - Útil para verificar a integridade e estrutura dos dados antes do pré-processamento.
````

````python
print(X.columns)  # 🏷️ Exibe o nome de todas as colunas (características) do DataFrame X.
                  #    - Essas colunas representam as 30 características extraídas dos tumores.
                  #    - Útil para conferir quais atributos estão disponíveis no conjunto de dados.
````

````python
y = data.target  # 🎯 Armazena os rótulos (labels) das amostras do conjunto de dados.
                 #    - `data.target` contém a classificação dos tumores:
                 #      - 0 → Tumor benigno.
                 #      - 1 → Tumor maligno.
                 #    - Esse vetor será a variável de saída (target) no modelo de machine learning.

print(y)  # 🔍 Exibe todos os valores da variável alvo.
          #    - Isso ajuda a verificar a distribuição das classes antes do treinamento do modelo.
````

````python
print(data.feature_names)  # 🏷️ Exibe o nome de todas as características do conjunto de dados Breast Cancer.
                           #    - Retorna uma lista com os nomes das 30 colunas (features) usadas para a classificação.
                           #    - Útil para entender quais atributos estão sendo analisados pelo modelo.
````

````python
print(data.target_names)  # 🏷️ Exibe os nomes das classes do conjunto de dados Breast Cancer.
                          #    - Retorna um array com os rótulos das classes:
                          #      - 'malignant' → Tumores malignos (1)
                          #      - 'benign' → Tumores benignos (0)
                          #    - Útil para interpretar os valores da variável `data.target`.
````

````python
print(X.shape)  # 📏 Exibe a dimensão do DataFrame X.
                #    - Retorna uma tupla (n_linhas, n_colunas).
                #    - No caso do Breast Cancer Dataset, o esperado é (569, 30):
                #      - 569 amostras (tumores analisados).
                #      - 30 características para cada amostra.
                #    - Útil para conferir o tamanho do conjunto de dados antes da modelagem.
````

````python
X_treino, X_teste, y_treino, y_teste = train_test_split(  # 🔀 Divide os dados em conjuntos de treino e teste.
    X,  # 📊 Conjunto de características (features), ou seja, os dados de entrada do modelo.
    y,  # 🎯 Conjunto de rótulos (labels), ou seja, a variável alvo que queremos prever.
    test_size=0.2,  # ✂️ Define que 20% dos dados serão usados para teste e 80% para treino.
    random_state=0,  # 🎲 Garante reprodutibilidade, mantendo a mesma divisão toda vez que o código for executado.
    stratify=y  # ⚖️ Mantém a proporção das classes (benigno/maligno) na divisão, evitando desbalanceamento nos conjuntos.
)
````

````python
print(X_treino.shape)  # 📏 Exibe a dimensão do conjunto de treino.
                        #    - Retorna (n_amostras, n_características).
                        #    - Esperado: (455, 30), já que 80% de 569 amostras foram alocadas para treino.

print(X_teste.shape)  # 📏 Exibe a dimensão do conjunto de teste.
                      #    - Retorna (n_amostras, n_características).
                      #    - Esperado: (114, 30), pois 20% das 569 amostras foram reservadas para teste.
````

````python
from sklearn.neighbors import KNeighborsClassifier  # 🤖 Importa o classificador K-Nearest Neighbors (KNN).
from sklearn.metrics import classification_report  # 📊 Importa a função para gerar um relatório de métricas de classificação.

clf = KNeighborsClassifier()  # 🎯 Cria um modelo KNN com os parâmetros padrão.
                              #    - O KNN classifica uma amostra com base na classe mais comum entre seus k vizinhos mais próximos.
                              #    - O valor padrão de k é 5.

clf.fit(X_treino, y_treino)  # 📚 Treina o modelo usando os dados de treino (X_treino e y_treino).

prediction = clf.predict(X_teste)  # 🔮 Faz previsões com os dados de teste.

print(clf.score(X_treino, y_treino))  # 📏 Exibe a acurácia do modelo no conjunto de treino.
                                       #    - Mede a proporção de previsões corretas sobre o total de amostras de treino.

print(clf.score(X_teste, y_teste))  # 📏 Exibe a acurácia do modelo no conjunto de teste.
                                     #    - Compara as previsões do modelo com os valores reais de y_teste.

print(classification_report(y_teste, prediction))  # 📜 Exibe um relatório detalhado com métricas da classificação:
                                                   #    - Precisão (Precision): Quantos dos positivos previstos são realmente positivos.
                                                   #    - Recall (Sensibilidade): Quantos dos positivos reais foram corretamente identificados.
                                                   #    - F1-score: Média harmônica entre precisão e recall (bom equilíbrio entre os dois).
                                                   #    - Suporte: Número de amostras de cada classe no conjunto de teste.
````

````python
escalonador = StandardScaler()  # ⚖️ Cria um escalonador padrão (StandardScaler) do Scikit-Learn.
                                #    - Ele padroniza os dados transformando-os para uma distribuição com média 0 e desvio padrão 1.
                                #    - Isso melhora o desempenho de algoritmos que são sensíveis à escala dos dados, como KNN e redes neurais.

X_treino = escalonador.fit_transform(X_treino)  # 📏 Ajusta o escalonador aos dados de treino e os transforma.
                                                #    - `fit_transform()` primeiro aprende os parâmetros de escalonamento (média e desvio padrão),
                                                #      depois aplica a transformação para padronizar os dados.

X_teste = escalonador.transform(X_teste)  # 🔄 Aplica a mesma transformação ao conjunto de teste.
                                          #    - Usa os mesmos parâmetros aprendidos com `X_treino` (média e desvio padrão),
                                          #      garantindo que os dados de treino e teste estejam na mesma escala.

````

````python
X_treino = X_treino.reshape(455, 30, 1)  # 🔄 Redimensiona os dados de treino para o formato (455, 30, 1).
                                         #    - `455` → Número de amostras no conjunto de treino.
                                         #    - `30` → Número de características por amostra.
                                         #    - `1`  → Adiciona uma dimensão extra para compatibilidade com redes neurais (ex.: CNNs).
                                         #    - Essa conversão pode ser necessária para modelos baseados em TensorFlow/Keras que esperam
                                         #      entradas em formato tridimensional.

X_teste = X_teste.reshape(114, 30, 1)  # 🔄 Redimensiona os dados de teste para o formato (114, 30, 1).
                                       #    - `114` → Número de amostras no conjunto de teste.
                                       #    - `30` → Número de características por amostra.
                                       #    - `1`  → Mantém a consistência com o conjunto de treino.
                                       #    - Isso permite que os dados de teste sejam processados corretamente pelo modelo treinado.
````

````python
modelo = Sequential()  # 🏗️ Cria um modelo sequencial no Keras, onde as camadas são adicionadas em sequência.

modelo.add(Conv1D(filters=16,  # 🛠️ Adiciona uma camada convolucional 1D com 16 filtros.
                  kernel_size=2,  # 🔍 O tamanho do kernel (janela de convolução) é 2.
                  activation='relu',  # ⚡ Usa a função de ativação ReLU para aumentar a não linearidade do modelo.
                  input_shape=(30, 1)))  # 📏 Define a forma da entrada como (30,1) → 30 características e 1 canal.

modelo.add(Dropout(0.2)),  # 🚀 Adiciona uma camada Dropout para evitar overfitting, descartando aleatoriamente 20% dos neurônios durante o treino.

modelo.add(Conv1D(filters=32,  # 🛠️ Adiciona outra camada convolucional 1D, agora com 32 filtros.
                  kernel_size=2,  # 🔍 O kernel continua com tamanho 2.
                  activation='relu'))  # ⚡ Ativação ReLU para aprendizado não linear.

modelo.add(Flatten())  # 📏 Achata a saída da convolução em um vetor 1D, preparando os dados para as camadas densas.

modelo.add(Dense(32,  # 🎯 Adiciona uma camada totalmente conectada (Densa) com 32 neurônios.
                 activation='relu'))  # ⚡ Ativação ReLU para capturar padrões mais complexos.

modelo.add(Dense(1,  # 🎯 Camada de saída com 1 neurônio (porque estamos prevendo uma classe binária).
                 activation='sigmoid'))  # 🔢 Ativação sigmoide, que transforma a saída em um valor entre 0 e 1 (probabilidade).

# 🔗 Link da documentação: https://keras.io/api/layers/convolution_layers/convolution1d/
````

````python
learning_rate = ReduceLROnPlateau(monitor='accuracy',   # 🔍 Monitora a métrica 'accuracy' (precisão) durante o treinamento.
                                  #    - Se a precisão parar de melhorar, a taxa de aprendizado será reduzida.

                                  factor=0.2,          # 📉 Reduz a taxa de aprendizado em 20% quando não houver melhoria.
                                  #    - A nova learning rate será a atual multiplicada por 0.2.

                                  patience=1,          # ⏳ Espera 1 época sem melhora antes de reduzir a taxa de aprendizado.
                                  #    - Evita reduzir a learning rate muito rapidamente.

                                  min_lr=0.000001,     # 🚧 Define um limite mínimo para a taxa de aprendizado.
                                  #    - A learning rate nunca será menor que 0.000001.

                                  verbose=1)           # 📢 Exibe mensagens no console quando a taxa de aprendizado for reduzida.
                                  #    - Útil para acompanhar ajustes automáticos durante o treinamento.

# 🔗 Documentação oficial: https://keras.io/api/callbacks/reduce_lr_on_plateau/
````

````python
# Métricas processadas durante o treinamento

acc = hist.history['accuracy']  # ✅ Acurácia ao longo das épocas.
                                #    - Mede a proporção de previsões corretas.

loss = hist.history['loss']  # ❌ Taxa de perda (Loss) ao longo do treinamento.
                             #    - Indica o erro do modelo em cada época.

fp = hist.history['fp']  # 🚨 Número de falsos positivos (FP).
                         #    - Casos em que o modelo previu "positivo", mas era "negativo".

fpv = hist.history['val_fp']  # 🚨 Falsos positivos no conjunto de validação.
                              #    - Mesmo conceito, mas avaliado nos dados de teste.

fn = hist.history['fn']  # ⚠️ Número de falsos negativos (FN).
                         #    - Casos em que o modelo previu "negativo", mas era "positivo".

fnv = hist.history['val_fn']  # ⚠️ Falsos negativos na validação.

tp = hist.history['tp']  # ✅ Número de verdadeiros positivos (TP).
                         #    - Casos em que o modelo acertou ao prever "positivo".

tpv = hist.history['val_tp']  # ✅ Verdadeiros positivos na validação.

tn = hist.history['tn']  # ✅ Número de verdadeiros negativos (TN).
                         #    - Casos em que o modelo acertou ao prever "negativo".

tnv = hist.history['val_tn']  # ✅ Verdadeiros negativos na validação.

# 🎯 Extração dos valores na última época de treinamento:

FP = hist.history['fp'][-1]  # 🚨 Nº de falsos positivos na última época.

FN = hist.history['fn'][-1]  # ⚠️ Nº de falsos negativos na última época.

TP = hist.history['tp'][-1]  # ✅ Nº de verdadeiros positivos na última época.

TN = hist.history['tn'][-1]  # ✅ Nº de verdadeiros negativos na última época.

LOSS = hist.history['loss'][-1]  # ❌ Taxa de perda na última época.

LOSSV = hist.history['val_loss'][-1]  # ❌ Taxa de perda na validação.

ACC = hist.history['accuracy'][-1]  # ✅ Acurácia na última época.

ACCV = hist.history['val_accuracy'][-1]  # ✅ Acurácia na validação.

PRE = hist.history['precision'][-1]  # 🎯 Precisão (Positive Predictive Value).
                                      #    - Mede a proporção de previsões "positivas" que realmente são corretas.

PREV = hist.history['val_precision'][-1]  # 🎯 Precisão na validação.

REC = hist.history['recall'][-1]  # 🔄 Revocação (Recall ou Sensibilidade).
                                  #    - Mede a capacidade do modelo de identificar os "positivos" corretamente.

RECV = hist.history['val_recall'][-1]  # 🔄 Revocação na validação.
LR = hist.history['learning_rate'][-1] # Taxa de aprendizado na última época de processamento
````

````python
# Métricas processadas após o final do treinamento

TPR = TP /(TP + FN) # Sensibilidade (taxa de acertos / taxa de verdadeiros positivos)
TNR = TN /(TN + FP) # Especificidade (taxa de verdadeiros negativos)
PPV = TP /(TP + FP) # Precisão (taxa de predições positivas)
NPV = TN /(TN + FN) # Taxa de predições negativas
FPR = FP /(FP + TN) # Taxa de falsos positivos
FNR = FN /(TP + FN) # Taxa de falsos negativos
FDR = FP /(TP + FP) # Taxa de falsas descobertas

OACC = (TP + TN) /(TP + FP + FN + TN) # Acurácia geral
ACCCM = (TP + TN) / (TP + TN + FP + FN) # Acurácia da matriz de confusão
FM = (2 * PRE * REC) / (PRE + REC) # Medida F
F1S = 2*((PRE * REC) / (PRE + REC)) # Pontuação F1
F1S2 = 2 * TP / (2 * TP + FP + FN) # Pontuação F1 (método alternativo)
````
