````python
METRICS = [  # ğŸ“Š Lista de mÃ©tricas usadas para avaliar o desempenho do modelo de machine learning.

    BinaryAccuracy(name='accuracy'),  # âœ… Mede a acurÃ¡cia do modelo em classificaÃ§Ãµes binÃ¡rias (0 ou 1).
                                      #    - Calcula a proporÃ§Ã£o de previsÃµes corretas sobre o total de amostras.

    TruePositives(thresholds=0.5, name='tp'),  # ğŸ¯ Conta quantas vezes o modelo classificou corretamente um caso positivo (1).
                                               #    - Exemplo: Classificar corretamente um paciente como doente.

    TrueNegatives(thresholds=0.5, name='tn'),  # âœ… Conta quantas vezes o modelo classificou corretamente um caso negativo (0).
                                               #    - Exemplo: Classificar corretamente um paciente como saudÃ¡vel.

    FalsePositives(thresholds=0.5, name='fp'),  # ğŸš¨ Conta os falsos positivos (quando o modelo previu 1, mas o correto era 0).
                                                #    - Exemplo: Diagnosticar um paciente saudÃ¡vel como doente.

    FalseNegatives(thresholds=0.5, name='fn'),  # âš ï¸ Conta os falsos negativos (quando o modelo previu 0, mas o correto era 1).
                                                #    - Exemplo: Diagnosticar um paciente doente como saudÃ¡vel.

    PrecisionAtRecall(recall=0.5, name='precision'),  # ğŸ¯ Mede a precisÃ£o do modelo para um nÃ­vel especÃ­fico de recall (sensibilidade).
                                                      #    - PrecisÃ£o = (TP) / (TP + FP).
                                                      #    - Ãštil para evitar falsos positivos em aplicaÃ§Ãµes crÃ­ticas.

    SensitivityAtSpecificity(0.5, name='sensitivity'),  # ğŸ”¬ Mede a sensibilidade (recall) para um valor especÃ­fico de especificidade.
                                                        #    - Sensibilidade = (TP) / (TP + FN).
                                                        #    - Ãštil para detectar o mÃ¡ximo de casos positivos.

    SpecificityAtSensitivity(sensitivity=0.5, name='specificity'),  # ğŸ›¡ï¸ Mede a especificidade para um nÃ­vel especÃ­fico de sensibilidade.
                                                                    #    - Especificidade = (TN) / (TN + FP).
                                                                    #    - Ãštil para minimizar falsos positivos.

    Recall(name='recall')  # ğŸ”„ Mede a capacidade do modelo de identificar corretamente os positivos.
                           #    - Recall = (TP) / (TP + FN).
                           #    - Importante para modelos onde perder um caso positivo Ã© crÃ­tico (exemplo: diagnÃ³stico de doenÃ§as).

]  # ğŸ”¢ Essa lista serÃ¡ usada ao compilar o modelo, permitindo acompanhar diferentes mÃ©tricas durante o treinamento e a avaliaÃ§Ã£o.
````

````python
X_treino, X_teste, y_treino, y_teste = train_test_split(  # ğŸ”€ Divide os dados em conjuntos de treino e teste.
    X,  # ğŸ“Š Conjunto de caracterÃ­sticas (features), ou seja, os dados de entrada do modelo.
    y,  # ğŸ¯ Conjunto de rÃ³tulos (labels), ou seja, a variÃ¡vel alvo que queremos prever.
    test_size=0.2,  # âœ‚ï¸ Define que 20% dos dados serÃ£o usados para teste e 80% para treino.
    random_state=0,  # ğŸ² Garante reprodutibilidade, mantendo a mesma divisÃ£o toda vez que o cÃ³digo for executado.
    stratify=y  # âš–ï¸ MantÃ©m a proporÃ§Ã£o das classes (benigno/maligno) na divisÃ£o, evitando desbalanceamento nos conjuntos.
)
````

````python
print(X_treino.shape)  # ğŸ“ Exibe a dimensÃ£o do conjunto de treino.
                        #    - Retorna (n_amostras, n_caracterÃ­sticas).
                        #    - Esperado: (455, 30), jÃ¡ que 80% de 569 amostras foram alocadas para treino.

print(X_teste.shape)  # ğŸ“ Exibe a dimensÃ£o do conjunto de teste.
                      #    - Retorna (n_amostras, n_caracterÃ­sticas).
                      #    - Esperado: (114, 30), pois 20% das 569 amostras foram reservadas para teste.
````

````python
from sklearn.neighbors import KNeighborsClassifier  # ğŸ¤– Importa o classificador K-Nearest Neighbors (KNN).
from sklearn.metrics import classification_report  # ğŸ“Š Importa a funÃ§Ã£o para gerar um relatÃ³rio de mÃ©tricas de classificaÃ§Ã£o.

clf = KNeighborsClassifier()  # ğŸ¯ Cria um modelo KNN com os parÃ¢metros padrÃ£o.
                              #    - O KNN classifica uma amostra com base na classe mais comum entre seus k vizinhos mais prÃ³ximos.
                              #    - O valor padrÃ£o de k Ã© 5.

clf.fit(X_treino, y_treino)  # ğŸ“š Treina o modelo usando os dados de treino (X_treino e y_treino).

prediction = clf.predict(X_teste)  # ğŸ”® Faz previsÃµes com os dados de teste.

print(clf.score(X_treino, y_treino))  # ğŸ“ Exibe a acurÃ¡cia do modelo no conjunto de treino.
                                       #    - Mede a proporÃ§Ã£o de previsÃµes corretas sobre o total de amostras de treino.

print(clf.score(X_teste, y_teste))  # ğŸ“ Exibe a acurÃ¡cia do modelo no conjunto de teste.
                                     #    - Compara as previsÃµes do modelo com os valores reais de y_teste.

print(classification_report(y_teste, prediction))  # ğŸ“œ Exibe um relatÃ³rio detalhado com mÃ©tricas da classificaÃ§Ã£o:
                                                   #    - PrecisÃ£o (Precision): Quantos dos positivos previstos sÃ£o realmente positivos.
                                                   #    - Recall (Sensibilidade): Quantos dos positivos reais foram corretamente identificados.
                                                   #    - F1-score: MÃ©dia harmÃ´nica entre precisÃ£o e recall (bom equilÃ­brio entre os dois).
                                                   #    - Suporte: NÃºmero de amostras de cada classe no conjunto de teste.
````

````python
escalonador = StandardScaler()  # âš–ï¸ Cria um escalonador padrÃ£o (StandardScaler) do Scikit-Learn.
                                #    - Ele padroniza os dados transformando-os para uma distribuiÃ§Ã£o com mÃ©dia 0 e desvio padrÃ£o 1.
                                #    - Isso melhora o desempenho de algoritmos que sÃ£o sensÃ­veis Ã  escala dos dados, como KNN e redes neurais.

X_treino = escalonador.fit_transform(X_treino)  # ğŸ“ Ajusta o escalonador aos dados de treino e os transforma.
                                                #    - `fit_transform()` primeiro aprende os parÃ¢metros de escalonamento (mÃ©dia e desvio padrÃ£o),
                                                #      depois aplica a transformaÃ§Ã£o para padronizar os dados.

X_teste = escalonador.transform(X_teste)  # ğŸ”„ Aplica a mesma transformaÃ§Ã£o ao conjunto de teste.
                                          #    - Usa os mesmos parÃ¢metros aprendidos com `X_treino` (mÃ©dia e desvio padrÃ£o),
                                          #      garantindo que os dados de treino e teste estejam na mesma escala.

````

````python
X_treino = X_treino.reshape(455, 30, 1)  # ğŸ”„ Redimensiona os dados de treino para o formato (455, 30, 1).
                                         #    - `455` â†’ NÃºmero de amostras no conjunto de treino.
                                         #    - `30` â†’ NÃºmero de caracterÃ­sticas por amostra.
                                         #    - `1`  â†’ Adiciona uma dimensÃ£o extra para compatibilidade com redes neurais (ex.: CNNs).
                                         #    - Essa conversÃ£o pode ser necessÃ¡ria para modelos baseados em TensorFlow/Keras que esperam
                                         #      entradas em formato tridimensional.

X_teste = X_teste.reshape(114, 30, 1)  # ğŸ”„ Redimensiona os dados de teste para o formato (114, 30, 1).
                                       #    - `114` â†’ NÃºmero de amostras no conjunto de teste.
                                       #    - `30` â†’ NÃºmero de caracterÃ­sticas por amostra.
                                       #    - `1`  â†’ MantÃ©m a consistÃªncia com o conjunto de treino.
                                       #    - Isso permite que os dados de teste sejam processados corretamente pelo modelo treinado.
````

````python
modelo = Sequential()  # ğŸ—ï¸ Cria um modelo sequencial no Keras, onde as camadas sÃ£o adicionadas em sequÃªncia.

modelo.add(Conv1D(filters=16,  # ğŸ› ï¸ Adiciona uma camada convolucional 1D com 16 filtros.
                  kernel_size=2,  # ğŸ” O tamanho do kernel (janela de convoluÃ§Ã£o) Ã© 2.
                  activation='relu',  # âš¡ Usa a funÃ§Ã£o de ativaÃ§Ã£o ReLU para aumentar a nÃ£o linearidade do modelo.
                  input_shape=(30, 1)))  # ğŸ“ Define a forma da entrada como (30,1) â†’ 30 caracterÃ­sticas e 1 canal.

modelo.add(Dropout(0.2)),  # ğŸš€ Adiciona uma camada Dropout para evitar overfitting, descartando aleatoriamente 20% dos neurÃ´nios durante o treino.

modelo.add(Conv1D(filters=32,  # ğŸ› ï¸ Adiciona outra camada convolucional 1D, agora com 32 filtros.
                  kernel_size=2,  # ğŸ” O kernel continua com tamanho 2.
                  activation='relu'))  # âš¡ AtivaÃ§Ã£o ReLU para aprendizado nÃ£o linear.

modelo.add(Flatten())  # ğŸ“ Achata a saÃ­da da convoluÃ§Ã£o em um vetor 1D, preparando os dados para as camadas densas.

modelo.add(Dense(32,  # ğŸ¯ Adiciona uma camada totalmente conectada (Densa) com 32 neurÃ´nios.
                 activation='relu'))  # âš¡ AtivaÃ§Ã£o ReLU para capturar padrÃµes mais complexos.

modelo.add(Dense(1,  # ğŸ¯ Camada de saÃ­da com 1 neurÃ´nio (porque estamos prevendo uma classe binÃ¡ria).
                 activation='sigmoid'))  # ğŸ”¢ AtivaÃ§Ã£o sigmoide, que transforma a saÃ­da em um valor entre 0 e 1 (probabilidade).

# ğŸ”— Link da documentaÃ§Ã£o: https://keras.io/api/layers/convolution_layers/convolution1d/
````

````python
modelo.summary()  # ğŸ“‹ Exibe um resumo da arquitetura do modelo.
                  #    - Mostra todas as camadas adicionadas ao modelo.
                  #    - Indica o tipo de camada, nÃºmero de parÃ¢metros e formato da saÃ­da de cada camada.
                  #    - Ãštil para verificar se a estrutura do modelo estÃ¡ correta antes do treinamento.
````

````python
camadas = dict([(layer.name, layer) for layer in modelo.layers])  # ğŸ“‚ Cria um dicionÃ¡rio chamado 'camadas', onde:
                                                                  #    - Cada chave Ã© o nome de uma camada do modelo (layer.name).
                                                                  #    - Cada valor Ã© a prÃ³pria camada correspondente (layer).
                                                                  #    - Isso Ã© feito percorrendo todas as camadas do modelo (modelo.layers) usando uma compreensÃ£o de lista.
                                                                  #    - O dict() converte essa lista de pares (nome, camada) em um dicionÃ¡rio.

print(f'NÂº de camadas: {len(camadas)}')  # ğŸ”¢ Exibe o nÃºmero total de camadas no modelo.
                                         #    - Usa len(camadas) para contar quantas camadas foram armazenadas no dicionÃ¡rio.
                                         #    - O f-string formata a saÃ­da para incluir o nÃºmero dentro da mensagem.
````

````python
tf.keras.backend.clear_session()  # ğŸ§¹ Limpa a sessÃ£o do backend do TensorFlow.
                                  #    - Libera a memÃ³ria ocupada por modelos e grÃ¡ficos computacionais anteriores.
                                  #    - Ajuda a evitar conflitos ao criar um novo modelo, especialmente em notebooks.
                                  #    - Ãštil para evitar consumo excessivo de RAM quando se testa vÃ¡rias arquiteturas.
````

````python
modelo.compile(optimizer=Adam(learning_rate=0.001),  # ğŸš€ Compila o modelo definindo o otimizador, a funÃ§Ã£o de perda e as mÃ©tricas.
               #    - optimizer=Adam(learning_rate=0.001): Usa o otimizador Adam para atualizar os pesos do modelo.
               #      - Adam Ã© um dos otimizadores mais populares para redes neurais.
               #      - O parÃ¢metro learning_rate=0.001 define a taxa de aprendizado, controlando o tamanho dos ajustes nos pesos.

               loss='binary_crossentropy',            # âš–ï¸ Define a funÃ§Ã£o de perda.
               #    - 'binary_crossentropy' Ã© usada para problemas de classificaÃ§Ã£o binÃ¡ria (duas classes, como 0 e 1).
               #    - Mede a diferenÃ§a entre as previsÃµes do modelo e os valores reais.

               metrics=METRICS)                       # ğŸ“Š Define as mÃ©tricas para avaliar o desempenho do modelo.
               #    - METRICS Ã© uma variÃ¡vel que contÃ©m as mÃ©tricas escolhidas (ex: precisÃ£o, recall, F1-score, etc.).
               #    - Essas mÃ©tricas serÃ£o monitoradas durante o treinamento para analisar a performance do modelo.
````

````python
learning_rate = ReduceLROnPlateau(monitor='accuracy',   # ğŸ” Monitora a mÃ©trica 'accuracy' (precisÃ£o) durante o treinamento.
                                  #    - Se a precisÃ£o parar de melhorar, a taxa de aprendizado serÃ¡ reduzida.

                                  factor=0.2,          # ğŸ“‰ Reduz a taxa de aprendizado em 20% quando nÃ£o houver melhoria.
                                  #    - A nova learning rate serÃ¡ a atual multiplicada por 0.2.

                                  patience=1,          # â³ Espera 1 Ã©poca sem melhora antes de reduzir a taxa de aprendizado.
                                  #    - Evita reduzir a learning rate muito rapidamente.

                                  min_lr=0.000001,     # ğŸš§ Define um limite mÃ­nimo para a taxa de aprendizado.
                                  #    - A learning rate nunca serÃ¡ menor que 0.000001.

                                  verbose=1)           # ğŸ“¢ Exibe mensagens no console quando a taxa de aprendizado for reduzida.
                                  #    - Ãštil para acompanhar ajustes automÃ¡ticos durante o treinamento.

# ğŸ”— DocumentaÃ§Ã£o oficial: https://keras.io/api/callbacks/reduce_lr_on_plateau/
````

````python
%%time
        #    - Exibe no console quanto tempo levou para treinar o modelo.


hist = modelo.fit(X_treino,                # ğŸ‹ï¸â€â™‚ï¸ Dados de treinamento (entradas).
                  y_treino,                # ğŸ¯ RÃ³tulos de treinamento (saÃ­das esperadas).

                  steps_per_epoch=10,      # ğŸ”„ NÃºmero de lotes (batches) processados por Ã©poca.
                                           #    - Define quantas iteraÃ§Ãµes sÃ£o feitas antes de considerar uma Ã©poca completa.

                  epochs=100,              # â³ Define o nÃºmero total de Ã©pocas do treinamento.
                                           #    - O modelo serÃ¡ treinado por 100 ciclos completos nos dados.

                  validation_data=(X_teste, y_teste),  # ğŸ§ª Conjunto de validaÃ§Ã£o para medir desempenho.
                                                       #    - Usa os dados de teste para avaliar o modelo apÃ³s cada Ã©poca.

                  validation_steps=10,     # ğŸ”¢ NÃºmero de lotes a serem usados na validaÃ§Ã£o por Ã©poca.
                                           #    - Controla quantas amostras de validaÃ§Ã£o sÃ£o avaliadas em cada ciclo.

                  callbacks=[learning_rate],  # ğŸ” Callback para ajuste da taxa de aprendizado.
                                              #    - Reduz automaticamente a learning rate caso a mÃ©trica monitorada pare de melhorar.

                  verbose=1)               # ğŸ“¢ Define o nÃ­vel de exibiÃ§Ã£o no console.
                                           #    - "1" exibe uma barra de progresso com mÃ©tricas de cada Ã©poca.
                                           #    - "0" oculta detalhes, "2" exibe apenas uma linha por Ã©poca.
````

````python
print(hist.history)  # ğŸ“œ Exibe todo o histÃ³rico do treinamento do modelo.
                     #    - O dicionÃ¡rio `hist.history` contÃ©m os valores de todas as mÃ©tricas registradas.
                     #    - Inclui mÃ©tricas como `accuracy`, `loss`, `precision`, `recall`, entre outras.
                     #    - Cada mÃ©trica Ã© armazenada como uma lista, onde cada posiÃ§Ã£o corresponde a uma Ã©poca.
                     #    - Ãštil para anÃ¡lise posterior e plotagem de grÃ¡ficos de desempenho.
````

````python
# MÃ©tricas processadas durante o treinamento

acc = hist.history['accuracy']  # âœ… AcurÃ¡cia ao longo das Ã©pocas.
                                #    - Mede a proporÃ§Ã£o de previsÃµes corretas.

loss = hist.history['loss']  # âŒ Taxa de perda (Loss) ao longo do treinamento.
                             #    - Indica o erro do modelo em cada Ã©poca.

fp = hist.history['fp']  # ğŸš¨ NÃºmero de falsos positivos (FP).
                         #    - Casos em que o modelo previu "positivo", mas era "negativo".

fpv = hist.history['val_fp']  # ğŸš¨ Falsos positivos no conjunto de validaÃ§Ã£o.
                              #    - Mesmo conceito, mas avaliado nos dados de teste.

fn = hist.history['fn']  # âš ï¸ NÃºmero de falsos negativos (FN).
                         #    - Casos em que o modelo previu "negativo", mas era "positivo".

fnv = hist.history['val_fn']  # âš ï¸ Falsos negativos na validaÃ§Ã£o.

tp = hist.history['tp']  # âœ… NÃºmero de verdadeiros positivos (TP).
                         #    - Casos em que o modelo acertou ao prever "positivo".

tpv = hist.history['val_tp']  # âœ… Verdadeiros positivos na validaÃ§Ã£o.

tn = hist.history['tn']  # âœ… NÃºmero de verdadeiros negativos (TN).
                         #    - Casos em que o modelo acertou ao prever "negativo".

tnv = hist.history['val_tn']  # âœ… Verdadeiros negativos na validaÃ§Ã£o.

# ğŸ¯ ExtraÃ§Ã£o dos valores na Ãºltima Ã©poca de treinamento:

FP = hist.history['fp'][-1]  # ğŸš¨ NÂº de falsos positivos na Ãºltima Ã©poca.

FN = hist.history['fn'][-1]  # âš ï¸ NÂº de falsos negativos na Ãºltima Ã©poca.

TP = hist.history['tp'][-1]  # âœ… NÂº de verdadeiros positivos na Ãºltima Ã©poca.

TN = hist.history['tn'][-1]  # âœ… NÂº de verdadeiros negativos na Ãºltima Ã©poca.

LOSS = hist.history['loss'][-1]  # âŒ Taxa de perda na Ãºltima Ã©poca.

LOSSV = hist.history['val_loss'][-1]  # âŒ Taxa de perda na validaÃ§Ã£o.

ACC = hist.history['accuracy'][-1]  # âœ… AcurÃ¡cia na Ãºltima Ã©poca.

ACCV = hist.history['val_accuracy'][-1]  # âœ… AcurÃ¡cia na validaÃ§Ã£o.

PRE = hist.history['precision'][-1]  # ğŸ¯ PrecisÃ£o (Positive Predictive Value).
                                      #    - Mede a proporÃ§Ã£o de previsÃµes "positivas" que realmente sÃ£o corretas.

PREV = hist.history['val_precision'][-1]  # ğŸ¯ PrecisÃ£o na validaÃ§Ã£o.

REC = hist.history['recall'][-1]  # ğŸ”„ RevocaÃ§Ã£o (Recall ou Sensibilidade).
                                  #    - Mede a capacidade do modelo de identificar os "positivos" corretamente.

RECV = hist.history['val_recall'][-1]  # ğŸ”„ RevocaÃ§Ã£o na validaÃ§Ã£o.
LR = hist.history['learning_rate'][-1] # Taxa de aprendizado na Ãºltima Ã©poca de processamento
````

````python
# MÃ©tricas processadas apÃ³s o final do treinamento

TPR = TP /(TP + FN) # Sensibilidade (taxa de acertos / taxa de verdadeiros positivos)
TNR = TN /(TN + FP) # Especificidade (taxa de verdadeiros negativos)
PPV = TP /(TP + FP) # PrecisÃ£o (taxa de prediÃ§Ãµes positivas)
NPV = TN /(TN + FN) # Taxa de prediÃ§Ãµes negativas
FPR = FP /(FP + TN) # Taxa de falsos positivos
FNR = FN /(TP + FN) # Taxa de falsos negativos
FDR = FP /(TP + FP) # Taxa de falsas descobertas

OACC = (TP + TN) /(TP + FP + FN + TN) # AcurÃ¡cia geral
ACCCM = (TP + TN) / (TP + TN + FP + FN) # AcurÃ¡cia da matriz de confusÃ£o
FM = (2 * PRE * REC) / (PRE + REC) # Medida F
F1S = 2*((PRE * REC) / (PRE + REC)) # PontuaÃ§Ã£o F1
F1S2 = 2 * TP / (2 * TP + FP + FN) # PontuaÃ§Ã£o F1 (mÃ©todo alternativo)
````
